```python
#Using Matplotlib
plt.figure(figsize = (10,7))
plt.title("Count plot of regions")
sns.countplot(x="Region",data = df_sales)
```

```python
#Using Matplotlib
plt.figure(figsize = (10,7))
plt.title("Count plot of regions")
sns.countplot(x="Region",data = df_sales)
plt.show()
```

```python
plt.figure(figsize = (10,7))
plt.title("Count plot of regions")
sns.countplot(x="Region",data = df_sales)
plt.xlabel("Different Regions")
plt.ylabel("Number Of Orders")
plt.xticks(rotation= 90)
plt.show()
```

```python
# Using Matplotlib
a = pd.DataFrame(df_sales['Region'].value_counts())
a
```

```python
a.columns
```

```python
a["Region"]
```

```python
plt.pie(a['Region'], labels = a.index, autopct = "%.f%%")
plt.show()
```

```python
#Using Seaborn with color palettes
colorss = sns.color_palette("pastel")[0:5]
plt.pie(a["Region"], labels = a.index,colors = colorss, autopct = "%.f%%")
plt.show()
```
